```python
import numpy as np
import pandas as pd
```

```python
# read the data from the excel sheet
movies=pd.read_csv('tmdb_5000_movies.csv')
credits=pd.read_csv('tmdb_5000_credits.csv')
```

```python
# merge two data set into one data set (common column is title)
movies=movies.merge(credits,on='title')
```

```python
# As a pre processing part we need to take some columns which will help to give more accurate output
# genres
# title
# cast
# crew
# keywords
# overview
# id
movies=movies[['genres','title','cast','crew','keywords','overview','id']]
```

```python
movies.isnull().sum()
```

```python
# drop the rows which has missing values
movies.dropna(inplace=True)
```

```python
import ast
```

```python
# this function will help to fetch the genre names as a list 
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name']) 
    return L 
```

```python
# update the genre column in dataframe with new genre
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['genres']
```

```python
# # this function will help to fetch the keywords names as a list 
movies['keywords'] = movies['keywords'].apply(convert)
movies.head()
```

```python
movies['keywords'][0]
```

```python
# this function will help to fetch first three actor/actress from the cast
def convert3(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    return L 
```

```python
# update the cast column in dataframe with new cast
movies['cast'] = movies['cast'].apply(convert3)
movies.head()
```

```python
movies['cast']
```

```python
movies['cast'] = movies['cast'].apply(lambda x:x[0:3])
```

```python
movies['cast']
```

```python
# this function helps to fetch the director name from crew list
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L  
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
movies.sample(5)
```

```python
# this function removes the space between two words
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
```

```python
movies.head()
```

```python
# convert the oveerview into list
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies['overview'][0]
```

```python
# create a new column named tags and add voerview, genres,keywords,cast,crew into the tags column
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
# drop the unnecessary columns
```

```python
new = movies.drop(columns=['overview','genres','keywords','cast','crew'])
```

```python
new.head()
```

```python
# convert the tags list into a string
```

```python
new['tags'] = new['tags'].apply(lambda x: " ".join(x))
new.head()
```

```python
from nltk.stem import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return  " ".join(y)   
```

```python
new['tags'] = new['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(new['tags']).toarray()
```

```python
# calculate the cosine distance bewteen one vector to other vectors
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity[1]
```

```python
# recommendation function
```

```python
def recommend(movie):
    index = new[new['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new.iloc[i[0]].title)
        
```

```python
recommend('Inception')
```
